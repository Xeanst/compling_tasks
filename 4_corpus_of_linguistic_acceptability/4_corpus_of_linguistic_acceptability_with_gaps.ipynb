Современные нейросетевые модели достигают высокой степени компетентности во многих прикладных задачах понимания естественного языка. ## Оценки приемлемости в лингвистике

Умение оценивать грамматическую правильность предложения на родном языке —  ключевое свойство языковой способности человека. Под грамматичностью подразумевается интуитивная оценка, основанная на врожденной грамматике человека. Помимо грамматической правильности, на общую приемлемость предложения влияют  факторы употребления: частотность слов, длина предложения, его сложность, семантическая сочетаемость слов, объем рабочей памяти говорящего.

- грамматическая правильность (well-formedness): цепочки слов, которые могут быть порождены некоторой (ментальной) грамматикой.

- приемлемость (acceptability): цепочки слов, которые носители языка признают в качестве правильных предложений своего языка.

Грамматическая правильность vs. семантическая осмысленность:

1. Colorless green ideas sleep furiously.
2. Furiously sleep ideas green colorless.


image.png


Суждения о приемлемости предложений имеют название «оценок приемлемости» (acceptability judgments). Они выступают как один из основных
типов данных для моделирования языковой способности людей и анализа восприятия говорящими определенного языкового выражения.

Для извлечения оценок приемлемости проводится синтаксический эксперимент, при котором исследователь явно спрашивает испытуемого, возможно ли определенное
языковое высказывание в языке.

Выделяют различные методы оценки приемлемости:

- нечисловые
  - выбор между альтернативами (forced-choice task)

  Респонденту представляется два предложения, из которых ему нужно
выбрать одно более  приемлемое. Методика позволяет эксплицитно сопоставить два условия и прямо ответить на качественный вопрос о том, есть ли между этими условиями различие.
  - категориальная оценка «да/нет» (yes-no task)
  
  Респондентам представляется одно предложение, о котором нужно сказать, является ли оно приемлемым или нет.
  
- численные
  - шкала Ликерта (Likert scale)

  Респонденту предоставляется пронумерованная шкала, крайние значения которой
соответствуют приемлемому или неприемлемому предложению. Респондент должен оценить предложение по данной шкале: обычно от 1 до 5 или от 1 до 7 (на них можно установить середину).

  - оценка величины стимула (magnitude estimation)

  Респонденту в качестве ориентира предлагается предложение, которому уже выставлена численная оценка (предложение-стандарт). Задача респондента — оценить предложение относительно стандарта.

[Schütze C., Sprouse J. (2014) Judgment Data. Research Methods in
Linguistics.](https://web.mit.edu/hackl/www/lab/turkshop/readings/schutze-sprouse2011.pdf)

## Корпус лингвистической приемлемости

Наиболее распространенным подходом для оценки языковой способности нейронных сетей является использование корпусов лингвистической приемлемости (Corpus of Linguistic Acceptability, CoLA). Впервые такой набор данных был предложен для английского языка исследователями в области лингвистики и компьютерных наук из Университа Нью-Йорка:

[Warstadt A., Singh A., Bowman S.R. (2019) Neural Network Acceptability Judgments.](https://arxiv.org/abs/1805.12471)

Автоматичекая оценка приемлемости определяется как задача бинарной классификации предложений. Приемлемые предложения содержат метку 1, неприемлемые — метку 0.

  [Ross J.R (1967) Constraints on Variables in Syntax.](https://dspace.mit.edu/handle/1721.1/15166)

CoLA делится на наборы данных внутри предметной области (in-domain) и вне предметной области (out-of-domain).
- Данные внутри области разделены на наборы для обучения (8551)/ разработку (527)/ тестирование (530), англ. train/ dev/ test. Все предложения взятые из одних и тех же 17 источников.
- Набор данных вне домена разделен на выборки для разработки (516) и тестирования (533). Примеры взяты из других 6 источников.

Разделение позволяет отслеживать два типа переобучения:
- подстраивание под конкретные предложения в обучающем наборе (внутри предметной области)
- подстраивание под конкретные источники и явления, представленные в обучающем наборе (вне предметной области)

### Выбор явлений и разметка

Задача оценки приемлемости требует выявления сложных грамматических контрастов. Успешная модель должна распознавать:
- морфологические аномалии, как несоответствие в глагольной формы (a)
- синтаксические аномалии, как *wh*-передвижение из сочиненной структуры (b)
- семантические аномалии, как нарушения требований одушевленности для аргументов (c)

Исключались примеры, относящиеся к четырем категориям:

- прагматически аномальные (d)

Поддаются интерпретации, но в странных сценариях, отличимых от правдоподобных сценариев только при наличии доступа к знаниям реального мира, не связанным с грамматикой.
- имеющие невозможное значение (e)

Часто используются для иллюстрации того, что предложение не может выразить определенный смысл. Этот пример может выразить только то, что кто-то, кроме Джона, любит Джона. Исключаются, поскольку не существует простого способа заставить классификатор приемлемости учитывать только рассматриваемую интерпретацию.
- нормативные правила (f)

Нарушают правила, которые обычно выучиваются явно, а не усваиваются естественным путем, и поэтому не считаются частью грамматических знаний носителя языка в лингвистической теории.
- несуществующие слова (g)

Слова, которые образованы регулярными словообразовательными способами, но отсутствуют в языке. Модели не смогут судить о приемлемости этих предложений, поскольку эти слова не будут фигурировать в их словарях.

### Метрики оценки

Расчитываются по матрице ошибок:
Снимок экрана 2023-02-10 143732.png

[Accuracy](https://scikit-learn.org/stable/modules/generated/sklearn.metrics.accuracy_score.html) — доля правильных ответов алгоритма среди всех ответов (непоказательна в задачах с неравными классами):

$$Accuracy = \frac{TP + TN}{TP + TN + FP + FN}$$


[Коэффициент корреляции Мэтьюса](https://scikit-learn.org/stable/modules/generated/sklearn.metrics.matthews_corrcoef.html) (Matthews Correlation Coefficient, MCC) — коэффициент корреляции между наблюдаемыми и предсказанными бинарными классификациями. Данная метрика используется для оценки качества бинарной классификации и позволяет учитывать дисбаланс классов. Принимает значение на интервале [-1, 1]. Значение 1 соответствует идеальному предсказанию, 0 — ситуации случайного предсказания, -1 — полностью противоположному предсказанию.

$$MCC = \frac{TP × TN - FP × FN}{\sqrt{(TP+FP)(TP+FN)(TN+FP)(TN+FN)}}$$

Пять аспирантов-лингвистов (носители английского языка) выполнили задание на оценку приемлемости предложений из CoLA. Среднее значение метрик: Acc=86,1%, MCC=0,697.

## Лингвистический анализ суждений о приемлемости

Помимо стандартной версии CoLA, доступной в составе бенчмарка GLUE, существует версия с грамматической разметкой. Она описана в статье тех же исследователей:

[Warstadt A., Bowman S.R. (2019) Linguistic Analysis of Pretrained Sentence Encoders with Acceptability Judgments.](https://arxiv.org/abs/1901.03438)

Данная версия содержит примеры из валидационной выборки (внутри и вне предметной области). Для каждого предложения указано, какое лингвистическое явление оно иллюстрирует. Разметка реализована в формате бинарных признаков: 1, если признак присутствует, 0 — если отсутствует. Было выделено 13 основных признаков и 59 второстепенных признаков.

Снимок экрана 2023-10-15 181333.png

- простые предложения с подлежащим, сказуемым и дополнением (**simple**: Bo ate an apple.)
-  конструкции с более сложной предикативной структурой (**predicate**)
  - конструкции с вспомогательным глаголом *be* (**copular constructions**: Bo is awake.)
  - малые клаузы (**small clauses**: I saw *Bo jump*)
  - результативы/ депиктивы (**resultatives/depictives**: Bo wiped *the table clean.*)
- различные типы адъюнктов (**adjunct**)
  - в именной группе (**NP adjunct**: The boy *with blue eyes* gasped.)
  - в глагольной группе (**VP adjunct**: Bo sang *with Jo*.)
  - обстоятельства времени (**temporal**: Bo awoke *at dawn*.)
  - обстоятельства места (**locative**: Bo jumped *on the bed*.)
- тип аргумента
  - косвенное дополнение (**obliques**: I gave a book *to Bo*.)
  - предложная группа (**PP arguments**: Bo voted *for Jones*.)
  - эксплетив (**expletives**: *It* seems that Bo left)
- аргументные альтернации (**argument alternations**)
  - добавление аргумента (**added arguments**: I baked *Bo* a cake)
  - удаление аргумента (**dropped arguments**: Bo *knows*.)
  - пассивный залог (**passive**: I was *applauded*.)
- связывание (**bind**)
  - связывание рефлексивов (**bound reflexives**: Bo loves *himself*.)
  - связывание других местоимений (**other bound pronouns**: Bo thinks *he* won.)
- *wh*-передвижение (**question**)
  - во вложенной клаузе (**embedded clause**: I know *who you are*.)
  - в матричной клаузе (**matrix clause**: Who are you?)
  - относительное придаточное (**relative clause**: Bo saw the guy *who left*)
  - нарушение островных ограничений (**island**: * What was *a picture of* hanging on the wall?)
  - эффект крысолова (**pied-piping**: *With whom* did you eat?)
- различные типы придаточных (**complement clause**)
  - субъектные (**subject CPs**: *That Bo won* is odd.)
  - сентенциальные актанты (**CP arguments of VPs or NPs/APs**: The fact *that Bo won*.)
  - без подчинительного союза (**no complementizer**: I think *Bo's crazy*.)
  - нефинитные (**non-finite CPs**: This is ready *for you to eat.*).
- вспомогательные глаголы (**auxiliary**)
  - наличие вспомогательного глагола (**auxiliary**)
  - наличие модального глагола (**modal**: I *can* win.)
  - отрицание (**negation**)
  - псевдовспомогательные глаголы (**pseudo-auxiliaries**: I *have* to win).
- различные типы инфинитивных глагольных групп (**to-VP**)
  - контроль (**control**: Bo wants *to win*.)
  - подъем (**raising**:  Bo seemed *to fly*.)
  - аргументы существительного и прилагательного (**VP arguments of NPs or APs**: Bo is eager *to eat*.)
  - глагольные группы с передвижением (**VPs with extraction**: This is easy *to read _*.)
- группы существительного и прилагательного (**noun, adjective**)
  - с предложными аргументами (**with PP arguments**: Bo is *fond of Mo*.)
  - присоединяют клаузу или глагольную группу (**CP/VP arguments**)
  - сложные слова (**noun-noun compounds**: Bo ate *mud pie*.)
  - образованные от глаголов (**derived from verbs**: *Baking* is fun.)
- другие синтаксические конструкции (**s-syntax**)
  - дислокация (**dislocated phrases**: The boy left *who was here earlier*)
  - передвижение, связанное с информационной структурой (**movement related to focus or information structure**: *This* I've gotta see _.)
  - сочиненные конструкции (**coordination**)
  - придаточные предложения (**subordinate clauses**)
  - эллипсис (**ellipsis**: I can't.)
  - адъюнкты на уровне предложения (**sentence level adjuncts**: *Apparently*, it's raining.)
- определители (**determiner**)
  - квантификаторы (**quantifiers**)
  - партитивные конструкции (**partitives**: two of the boys)
  - единицы отрицательной полярности (**negative polarity items**: I *do/don't
have *any* pie)
  - сравнительные конструкции (**comparative constructions**)
